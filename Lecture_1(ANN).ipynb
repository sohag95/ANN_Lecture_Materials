## **Perceptron: The Fundamental Building Block of Neural Networks**

Perceptron is the fundamental unit of ANN that processes input, applies weights and bias, and passes the result through an activation function.

| **Building Block** | **Explanation** |
| ----- | ----- |
| **Inputs** | The features or data points fed into the perceptron. |
| **Weights (W)** | Coefficients assigned to inputs, determining their importance. |
| **Bias (b)** | A constant term that helps adjust the decision boundary. |
| **Summation (∑)** | Computes the weighted sum of inputs: ∑(Wi.Xi)+b  |
| **Activation Function** | Applies a threshold (e.g., step function) to produce the output. |
| **Output** | The final decision (0 or 1) based on the activation function. |

### **Perceptron Training Algorithm**
The following algorithm outlines the steps involved in training a perceptron, where weights are adjusted iteratively based on errors to achieve accurate classification.

**The process typically follows these steps:**

* **Initialize weights:** Set initial weights randomly.

* **Forward pass:** Compute the weighted sum of inputs and pass it through an activation function.

* **Calculate error:** Compare the predicted output with the actual target.

* **Update weights:** Adjust weights based on the error using a learning rate and the gradient descent algorithm.

* **Repeat:** Iterate through the dataset multiple times(epochs) until the more converges.

### **Perceptron Trick (Weight Update Rule)**
The Perceptron trick is a simple learning rule that updates the weights whenever the perceptron misclassifies a data point. The goal is to adjust the weights in such a way that the perceptron correctly classifies all training examples.

**Weight Update Rule:**

If a given input **x** is misclassified, update the weights as follows:

* **W=W+ΔW**

Where the weight update **Δ𝑊** is calculated as:

* **ΔW=η⋅(y_true−y_predicted)⋅x**

Similarly, the bias is updated as:

* **𝑏=𝑏+𝜂⋅(𝑦_true − 𝑦_predicted)**


**Explanation of Terms:**

* **W:** Weight vector

* **η (learning rate):** A small positive value that controls the step size of the update

* **y_true:** Actual label (0 or 1)

* **y_predicted:** Output predicted by the perceptron

* **x:** Input feature vector

* **b:** Bias term


### **How It Works:**
* If the perceptron correctly classifies an input, no weight update is needed.
* If the perceptron misclassifies an input:

* * **If it predicts 0 but should be 1**, increase the weights in the direction of
x.
* * **If it predicts 1 but should be 0**, decrease the weights in the direction of x.

* This iterative adjustment continues until all training examples are correctly classified or a stopping condition is met.

This weight update rule ensures that the perceptron gradually learns a **decision boundary** that separates different classes in the dataset.



## **Implementing Perceptron Training in Code**

Now that we understand the perceptron learning rule, let's implement it in Python. We will write a simple perceptron training algorithm that adjusts the weights iteratively based on misclassified examples. This implementation will help us see how the perceptron learns a decision boundary from data.

```python
import numpy as np
import numpy as np
from sklearn.metrics import confusion_matrix

# Define activation function
def activation_function(z):
    return 1 if z >= 0 else 0

# Perceptron Learning Algorithm
def perceptron_train(X, y, learning_rate=0.1, epochs=10):
    n_samples = X.shape[0]
    n_features = X.shape[1]

    # 1.Initialize weights
    weights = np.zeros(n_features)
    bias = 0

    for epoch in range(epochs):
        for i in range(n_samples):

            # 2.Forward pass
            z = np.dot(X[i], weights) + bias
            #z=(X[i][0]*weights[0])+(X[i][1]*weights[1])+bias
            y_pred = activation_function(z)

            # 3.Compute error
            error = y[i] - y_pred

            # Update Weights (Perceptron Trick)
            weights += learning_rate * error * X[i]
            # weights[0] += learning_rate * error * X[i][0]
            # weights[1] += learning_rate * error * X[i][1]
            bias += learning_rate * error

    return weights, bias
```

### **Perceptron Prediction Results for AND logic gate**

```python
# Example dataset (AND logic gate)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])  # AND function

# Train perceptron
weights, bias = perceptron_train(X, y)

# Print final weights
print("Final weights:", weights)
print("Final bias:", bias)
```

```python
# Test the trained model
for x, label in zip(X, y):
    prediction = activation_function(np.dot(x, weights) + bias)
    print(f"Input: {x}, Predicted: {prediction}, Actual: {label}")
```

```python
import matplotlib.pyplot as plt
w1, w2 = weights[0],weights[1]  # Extract learned weights
b = bias  # Extract bias

# Plot decision boundary
x_vals = np.linspace(-0.5, 1.5, 100)
y_vals = -(w1 * x_vals + b) / w2  # Solve for y in w1*x1 + w2*x2 + b = 0

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.plot(x_vals, y_vals, 'k--', label="Decision Boundary")
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.title('Perceptron Decision Boundary')
plt.legend()
plt.show()
```

## **Limitations of Perceptron**

* The **perceptron works well for linearly separable problems** (e.g., AND, OR gates).

* **Fails on non-linearly separable problems** like XOR, as it can only learn a **single linear decision boundary**.

* This limitation shows that a **single-layer perceptron is not enough** for complex classification tasks.

## **Next Step: Multi-Layer Perceptron (MLP)**
To overcome the perceptron's limitations, we introduce the **Multi-Layer Perceptron (MLP):**

* Uses multiple layers (input, hidden, and output layers).

* Employs non-linear activation functions to handle complex decision boundaries.

* Trains using backpropagation, enabling effective learning from data.

MLP allows us to solve XOR and other non-linearly separable problems, making it a powerful model for deep learning.
